## Data Splitting

```python
x=data.iloc[:,:-1]
```

```python
y=data.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=13)
```

## Model Creation

```python
reg1=LogisticRegression()
reg2=DecisionTreeClassifier(max_depth=4)
reg3=SVC()
```

```python
voting_clas=VotingClassifier(estimators=[('LR',reg1),('DT',reg2),("SVC",reg3)])
```

```python
voting_clas.fit(x_train, y_train)
```

```python
y_pred=voting_clas.predict(x_test)
```

## Evaluatng

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

## Hard and Soft voting

```python
hard_voting=VotingClassifier(estimators=[('LR',reg1),('DT',reg2),("SVC",reg3)], voting='hard')
hard_voting.fit(x_train,y_train)
hard_voting_pred=hard_voting.predict(x_test)
print(classification_report(y_test,hard_voting_pred))
```

```python
soft_voting=VotingClassifier(estimators=[('LR',reg1),('DT',reg2)], voting='soft')
soft_voting.fit(x_train,y_train)
soft_voting_pred=soft_voting.predict(x_test)
print(classification_report(y_test,soft_voting_pred))
```

## Weighted Voting

```python
voting_clas=VotingClassifier(estimators=[('LR',reg1),('DT',reg2),("SVC",reg3)],weights=[0.2,0.5,0.3])
```

```python
voting_clas.fit(x_train, y_train)
```

```python
WG_y_pred=voting_clas.predict(x_test)
```

```python
print(classification_report(y_test,WG_y_pred))
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
tree=DecisionTreeClassifier(max_depth=4)
```

```python
tree.fit(x_train, y_train)
```

```python
Bagging=BaggingClassifier(estimator=tree,n_estimators=5,max_samples=50,bootstrap=True,random_state=13)
Bagging.fit(x_train, y_train)
```

```python
print(classification_report(y_test,Bagging.predict(x_test)))
```
