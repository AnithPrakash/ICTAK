```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Age",y="Glucose",data=data)
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.title("Age Vs Glucose count")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x="Outcome",y="Age",data=data)
plt.xlabel("Outcome")
plt.ylabel("Age")
plt.title("Age distribution")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="Insulin",y="Glucose",data=data)
plt.xlabel("Insulin")
plt.ylabel("Glucose")
plt.title("Insulin Usage Vs Glucose count")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.hist(data['BMI'],color="b",edgecolor="r")
plt.xlabel("BMI")
plt.show()
```
